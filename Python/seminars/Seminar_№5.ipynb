# Задача 2

Постройте столбчатую диаграмму по количеству вакцинированных  в каждой стране.

```python
!wget -O "vaccinations.csv" "https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/vaccinations/vaccinations.csv"
```

## Решение

### Описание данных:
- location - Страна
- iso_code - Сокращенное название страны (код страны)
- date - Дата
- total_vaccinations - Общее количество вакцинированных

```python
import matplotlib.pyplot as plt
```

#### Открываем файл

```python
location = []
total_vaccinations = []
file= open("vaccinations.csv", "r")
for line in file:
    location.append(line.split(',')[0])
    total_vaccinations.append(line.split(',')[3])
file.close()
```

#### Удаляем заголовки списков

```python
location.remove('location')
total_vaccinations.remove('total_vaccinations')
```

#### Посмотрим на список стран 

```python
country_list = list(set(location))
len(country_list)
```

#### Заменим пустые строки на нулевые значения, чтобы далее можно было работать со значениями этого списка, как с числами 

```python
e = []
for i in total_vaccinations:
    if i is "": 
        e.append("0")
    else:
        e.append(i)

total_vaccinations = e
#print(total_vaccinations)
```

#### Приведем все числа в списке total_vaccinations к формату int

```python
total_vaccinations = list(map(int, total_vaccinations))
```

#### Соединим два столбца location и total_vaccinations в общий массив 

```python
zipped = zip(location, total_vaccinations)
zl = list(zipped)
```

#### Получим максимальное значение - общее количество вакцинированных для каждой из стран

```python
data = [(location, total_vaccinations) for location, total_vaccinations in dict(sorted(zl)).items()]
data_sorted = sorted(data, key=lambda x: x[1], reverse=True)
```

#### Так как мы хотим получить статистику вакцинированных по странам, удалим из списка элементы, которые характеризуют не страны, а целые континенты

```python
labels=['World', 'Asia', 'Upper middle income', 'High income', 'Lower middle income', 'Europe', 'North America', 'European Union', 'South America', 'Africa']
```

```python
data_sorted_clear = [(c,v) for (c,v) in data_sorted if c not in labels]
```

#### Получим значения топ-10 стран по количеству вакцинированных

```python
top_10 = data_sorted_clear[:10]
top_10
```

#### Представим полученные данные в графическом виде:

```python
plt.bar(*zip(*top_10))
plt.ylabel('Общее количество вакцинированных')
plt.xlabel('Страна')
plt.title('Топ-10 стран по количеству вакцинированных')
plt.xticks(rotation=90)
plt.show()
```

Вывод: наибольшее количество вакцинированных в таких странах, как Китай, Индия и США. В остальных странах общее количество вакцинированных существенно ниже. Россия находится в этом списке на 9-ом месте (18.12.2021).
